```python
import pandas as pd

# read the dataframes
input2 = 'data/FingerprintTest.csv'
input1 = 'data/ReactionSmilesTest.csv'
output= 'data/output_file.csv'


df2 = pd.read_csv(input2, sep=',')
df1 = pd.read_csv(input1, sep=',')

    
# Crear un diccionario que asocie cada SMILES con sus respectivos descriptores
descriptors_dict = df2.set_index('SMILES').to_dict('index')
    
# Crear un nuevo dataframe para las reacciones con columnas adicionales para los descriptores
new_reaction_df = pd.DataFrame(columns=['Reaction SMILES'])
    
# Iterar a través de cada fila de la tabla de reacciones
for index, row in df1.iterrows():
    reaction_smiles = row['Reaction']
    substrates, products = reaction_smiles.split('>')
    substrate_descriptors = []
    product_descriptors = []
        
        # Buscar los descriptores correspondientes para cada molécula
    for substrate in substrates.split('.'):
        if substrate in descriptors_dict:
            substrate_descriptors.append(descriptors_dict[substrate])
    for product in products.split('.'):
        if product in descriptors_dict:
            product_descriptors.append(descriptors_dict[product])
        
    # Combinar los descriptores para cada reacción
    combined_descriptors = {}
    for substrate_descriptor in substrate_descriptors:
        for key, value in substrate_descriptor.items():
            combined_descriptors[key] = value
    for product_descriptor in product_descriptors:
        for key, value in product_descriptor.items():
            combined_descriptors[key] = value
        
        # Agregar la nueva fila al dataframe de reacciones
    new_row = {'Reaction': reaction_smiles}
    new_row.update(combined_descriptors)
    new_reaction_df = new_reaction_df.append(new_row, ignore_index=True)
    
# Escribir el nuevo dataframe como un archivo CSV
new_reaction_df.to_csv(output)
    
print(new_reaction_df)
```
